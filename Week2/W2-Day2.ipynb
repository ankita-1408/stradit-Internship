```python
#write lines (path, lines)
def write_lines(lines):

    with open("random.txt", "w") as file:

        for line in lines:
            file.write(line + "\n")


write_lines([
    "Hello",
    "Learning Python",
    "Day 2"
])
```

```python
#Count Non empty Lines
def count_lines():
    try:
        with open("random.txt", "r") as file:
            count = 0

            for line in file:
                if line.strip():
                    count += 1

            return count

    except FileNotFoundError:

        return 0

print(count_lines())
```

```python
#Count Words

def word_count():

    with open("random.txt", "r") as file:
        content = file.read()

    return len(content.split())

print(word_count())
```

```python
#Tail (last n line)
def tail(n):

    with open("random.txt", "r") as file:

        lines = file.readlines()

    return lines[-n:]

print(tail(2))
```

```python
#ADULT CSV
import csv


with open(
    "people.csv",
    "r"
) as file:

    reader = csv.DictReader(file)

    adults = []

    for row in reader:

        if int(row["age"]) >= 30:

            adults.append(row)


with open(
    "adults.csv",
    "w",
    newline=""
) as file:

    writer = csv.DictWriter(
        file,
        fieldnames=[
            "name",
            "age",
            "city"
        ]
    )

    writer.writeheader()

    writer.writerows(adults)


print("adults.csv created")
```

```python
#calculate average
import csv


total = 0

count = 0


with open(
    "scores.csv",
    "r"
) as file:

    reader = csv.DictReader(file)

    for row in reader:

        total += float(
            row["score"]
        )

        count += 1


print(
    "Average Score:",
    total/count
)
```

```python
#List to Dict
import csv


def csv_to_dict_list(
    path
):

    with open(
        path,
        "r"
    ) as file:

        reader = csv.DictReader(
            file
        )

        return list(
            reader
        )


result = csv_to_dict_list(
    "people.csv"
)


print(result)
```

```python
#list Dict to CSV
import csv


def dict_list_to_csv(
    records,
    path
):

    fields = (
        records[0]
        .keys()
    )


    with open(
        path,
        "w",
        newline=""
    ) as file:

        writer = csv.DictWriter(
            file,
            fieldnames=fields
        )

        writer.writeheader()

        writer.writerows(
            records
        )


data = [

{
"name":"Ankita",
"age":22,
"city":"Pune"
},

{
"name":"Amit",
"age":30,
"city":"Mumbai"
}

]


dict_list_to_csv(
    data,
    "new_people.csv"
)


print(
    "new_people.csv created"
)
```

```python
#Safe Integer
def safe_int(s, default=0):

    try:
        return int(s)

    except ValueError:
        return default


print(safe_int("100"))
print(safe_int("abc"))
```

```python
#Ask until file exist
def open_file():

    while True:

        name = input("Enter filename: ")

        try:

            with open(name, "r") as file:

                print("File opened successfully")

                print(file.read())

                break

        except FileNotFoundError:

            print("File not found")


open_file()
```

```python
#Safe Divides
def divide(a, b):

    try:
        return a / b

    except TypeError:
        return None

    except ZeroDivisionError:
        return None


print(divide(10,2))
print(divide(10,0))
print(divide("a",2))
```

```python
#Read Numbers and skip invalid
def read_numbers(path):

    valid = []
    skipped = 0

    try:

        with open(path) as file:

            for line in file:

                try:
                    valid.append(float(line))

                except ValueError:
                    skipped += 1

        print("Numbers:", valid)
        print("Skipped:", skipped)

    except FileNotFoundError:
        print("File not found")


read_numbers("random.txt")
```

```python
#Custom exception

class EmptyFileError(Exception):
    pass


def check_file(path):

    with open(path, "r") as file:

        content = file.read()

        if content == "":
            raise EmptyFileError(
                "File is empty"
            )

    return content


try:

    output = check_file("random.txt")   

    print(output)

except EmptyFileError as e:

    print(e)
```

```python
#create safe conversion
def safe_number(
    value,
    row,
    column
):

    try:

        return float(
            value
        )

    except ValueError:

        print(
            f"Warning → Row {row}, Column {column} converted to 0"
        )

        return 0
```

```python
#calculate total + average
import csv


result = []


with open(
    "students.csv",
    "r"
) as file:


    reader = csv.DictReader(
        file
    )


    for row_num, row in enumerate(
        reader,
        start=2
    ):


        math = safe_number(
            row["math"],
            row_num,
            "math"
        )

        science = safe_number(
            row["science"],
            row_num,
            "science"
        )

        english = safe_number(
            row["english"],
            row_num,
            "english"
        )


        total = (
            math
            + science
            + english
        )

        average = (
            total / 3
        )


        row["total"] = total

        row["average"] = round(
            average,
            2
        )


        result.append(
            row
        )


print(result)
```

```python
#read and process
import csv


result = []


try:

    with open(
        "students.csv",
        "r"
    ) as file:


        reader = csv.DictReader(
            file
        )


        for row_num, row in enumerate(
            reader,
            start=2
        ):


            math = safe_number(
                row["math"],
                row_num,
                "math"
            )

            science = safe_number(
                row["science"],
                row_num,
                "science"
            )

            english = safe_number(
                row["english"],
                row_num,
                "english"
            )


            total = (
                math
                + science
                + english
            )


            average = (
                total / 3
            )


            row["total"] = total

            row["average"] = round(
                average,
                2
            )


            result.append(
                row
            )


    print(
        "Data processed"
    )


except FileNotFoundError:

    print(
        "Error: students.csv is missing"
    )
    
```
