```python
#line chart

import matplotlib.pyplot as plt

x = list(range(-10, 11))

y = [i**2 for i in x]

plt.plot(x, y, marker='o')

plt.title("Graph of y = x^2")
plt.xlabel("x values")
plt.ylabel("y values")

plt.grid(True)
plt.show()
```

```python
#plot sin(x) and cos(x)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 500)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)")

plt.title("Sine and Cosine Waves")
plt.xlabel("X values")
plt.ylabel("Function Values")

plt.legend()

plt.grid(True)

plt.show()
```

```python
#Training Loss
import numpy as np
import matplotlib.pyplot as plt

epochs = np.arange(1, 21)

loss = 1.5 * np.exp(-0.1 * epochs) + np.random.normal(0, 0.03, 20)

plt.plot(
    epochs,
    loss,
    marker='o',
    linestyle='--',
    label='Training Loss'
)

# Add title and labels
plt.title("Training Loss Over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")

plt.grid(True)

plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("titanic.csv")

# Fill missing Age with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Embarked with mode
df["Embarked"] = df["Embarked"].fillna(
    df["Embarked"].mode()[0]
)

# Fill missing Fare with median
df["Fare"] = df["Fare"].fillna(
    df["Fare"].median()
)

# Family size
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

# IsAlone feature
df["IsAlone"] = np.where(
    df["FamilySize"] == 1,
    1,
    0
)

# Create Age Groups (5-year bins)
bins = range(0, 85, 5)

df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=bins
)

# Average fare by age group
avg_fare_by_age = df.groupby(
    "AgeGroup",
    observed=False
)["Fare"].mean()

# Create Figure
plt.figure(figsize=(18, 5))


# Chart 1 — Survival by Pclass
plt.subplot(1, 3, 1)

survival_by_class = df.groupby(
    "Pclass"
)["Survived"].mean()

plt.bar(
    survival_by_class.index.astype(str),
    survival_by_class.values,
    color=["red", "blue", "green"]
)

plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")


# Chart 2 — Age Distribution
plt.subplot(1, 3, 2)

plt.hist(
    df["Age"],
    bins=20,
    color="orange"
)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

# Chart 3 — Average Fare by Age Group

plt.subplot(1, 3, 3)

plt.plot(
    avg_fare_by_age.index.astype(str),
    avg_fare_by_age.values,
    marker='o'
)

plt.title("Average Fare by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Fare")

# Rotate labels for readability
plt.xticks(rotation=90)


plt.tight_layout()


# Save Figure
plt.savefig(
    "week2_deliverable.png",
    dpi=300,
    bbox_inches="tight")

plt.show()

# 100-word Summary
print("""
The first chart shows survival rate by passenger class.
Passengers in higher classes had better survival rates,
indicating socioeconomic advantage during evacuation.

The second chart displays the distribution of passenger ages.
Most passengers were young adults, with fewer elderly passengers.

The third chart shows the average fare paid across different
age groups. Some middle-age groups paid higher fares on average,
possibly because wealthier passengers tended to travel in certain
age ranges.

Together, these charts provide insight into survival patterns,
age distribution, and fare trends in the Titanic dataset using
data visualization techniques learned during Week 2.
""")
```
