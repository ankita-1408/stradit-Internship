```python
#Pangram Checker
#A pangram contains every letter a–z at least once.
def is_pangram(sentence):

    letters = set("abcdefghijklmnopqrstuvwxyz")

    return set(sentence.lower()).issuperset(letters)


text = input("Enter sentence: ")

print(is_pangram(text))
```

```python
#build a dictionary mapping each word to its frequency
sentence = input("Enter sentence: ")

words = sentence.lower().split()

freq = {}

for word in words:

    if word in freq:

        freq[word] += 1

    else:

        freq[word] = 1


print(freq)
```

```python
#Calculate total transaction amount per user
transactions = [

{"user":"a","amt":10},

{"user":"b","amt":7},

{"user":"a","amt":3}

]


totals = {}


for t in transactions:

    user = t["user"]

    amount = t["amt"]


    if user in totals:

        totals[user] += amount

    else:

        totals[user] = amount


print(totals)
```

```python
#SMART MINI STORE

# MINI E-COMMERCE INVENTORY

inventory = [

{
"sku": "A1",
"name": "T-shirt",
"price": 499,
"tags": ["clothing", "cotton"],





"stock": 12
},

{
"sku": "B2",
"name": "Mug",
"price": 199,
"tags": ["kitchen", "ceramic"],
"stock": 0
},

{
"sku": "C3",
"name": "Hoodie",
"price": 1299,
"tags": ["clothing", "winter"],
"stock": 4
},

{
"sku": "D4",
"name": "Notebook",
"price": 99,
"tags": ["stationery"],
"stock": 50
}

]

print("\n===== MY STORE DASHBOARD =====")

# 1. Items currently in stock

print("\nAVAILABLE PRODUCTS\n")

in_stock = [

item

for item in inventory

if item["stock"] > 0

]

for item in in_stock:

    print(
        item["name"],
        "| ₹", item["price"],
        "| Stock:", item["stock"]
    )

# 2. Most expensive item

expensive = max(
    inventory,
    key=lambda x: x["price"]
)

print("\nMOST PREMIUM PRODUCT")

print(
    expensive["name"],
    "- ₹",
    expensive["price"]
)

# 3. Average price


avg_price = (

sum(
item["price"]

for item in inventory
)

/

len(inventory)

)

print("\nAVERAGE PRODUCT PRICE")

print("₹", round(avg_price, 2))


# 4. Unique tags

all_tags = {

tag

for item in inventory

for tag in item["tags"]

}

print("\nSTORE CATEGORIES")

print(sorted(all_tags))

# 5. Tag → Products


tag_map = {}

for item in inventory:

    for tag in item["tags"]:

        if tag not in tag_map:

            tag_map[tag] = []

        tag_map[tag].append(
            item["name"]
        )


print("\nCATEGORY MAPPING")

for tag, products in tag_map.items():

    print(
        tag,
        "→",
        products
    )

# 6. Add sale badge


for item in inventory:

    if item["price"] < 500:

        item["on_sale"] = True

    else:

        item["on_sale"] = False


print("\nFINAL INVENTORY\n")

for item in inventory:

    print(item)
```
