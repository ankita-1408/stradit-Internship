```python
# Passenger Profile Report

import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)


# Fill missing Age with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Embarked with mode
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop Cabin column
df = df.drop(columns=["Cabin"])

# Extract Title from Name
df["Title"] = (
    df["Name"]
    .str.split(", ").str[1]
    .str.split(".").str[0]
)

# Create FamilySize
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

# Create IsAlone
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)

# Create FareCategory
df["FareCategory"] = pd.cut(
    df["Fare"],
    bins=[0, 10, 50, 100, 600],
    labels=["low", "mid", "high", "premium"]
)

# Survival Analysis

print("Survival Rate by Title:")
print(df.groupby("Title")["Survived"].mean())

print("\nSurvival Rate by Family Size:")
print(df.groupby("FamilySize")["Survived"].mean())

print("\nSurvival Rate by Fare Category:")
print(df.groupby("FareCategory", observed=False)["Survived"].mean())

# Save Cleaned Dataset
df.to_csv("titanic_clean.csv", index=False)

print("\nCleaned dataset saved as titanic_clean.csv")

# 5-Bullet Summary
female_rate = df[df["Sex"] == "female"]["Survived"].mean()

male_rate = df[df["Sex"] == "male"]["Survived"].mean()

first_class_rate = df[df["Pclass"] == 1]["Survived"].mean()
third_class_rate = df[df["Pclass"] == 3]["Survived"].mean()

alone_rate = df[df["IsAlone"] == 1]["Survived"].mean()
family_rate = df[df["IsAlone"] == 0]["Survived"].mean()

print("\nOBSERVATIONS")

print(f"1. Females had survival rate of {female_rate:.2f} vs males' {male_rate:.2f}")

print(f"2. 1st class passengers had survival rate of {first_class_rate:.2f} vs 3rd class passengers' {third_class_rate:.2f}")

print(f"3. Passengers travelling with family had survival rate of {family_rate:.2f} vs alone passengers' {alone_rate:.2f}")

print("4. Premium fare passengers generally had higher survival chances.")

print("5. Titles like 'Mrs' and 'Miss' showed better survival rates than 'Mr'.")
```
