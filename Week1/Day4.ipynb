```python
# Circle Calculator

PI = 3.14159

def circle_area(radius):
    return PI * radius ** 2


def circle_circumference(radius):
    return 2 * PI * radius


radii = [1, 5, 10]

print("Circle Calculator")

for r in radii:
    area = circle_area(r)
    circumference = circle_circumference(r)

    print(f"Radius: {r}")
    print(f"Area: {area:.2f}")
    print(f"Circumference: {circumference:.2f}")
    print("")
```

```python
#Pallindrom Checker

def is_pallindrome(text):
    text = text.lower()
    return text == text[::-1]

print("Pallindrome Checker")

words = ["Madam", "racecar", "hello", "Civic"]

for word in words:
    print(f"{word}: {is_pallindrome(word)}") 

print()
```

```python
#Password Generator
import random
import string

def generate_password(length=12):
    characters = string.ascii_letters + string.digits + string.punctuation

    password = ""

    for _ in range(length):
        password += random.choice(characters)
        
    return password

print("Password Generator")

for i in range(5):
    print(generate_password())
```

```python
#Sentence Analyser

def sentence_analyser(text):
    vowel = "aeiou"

    words = text.split()

    word_count = len(words)

    char_count = len(text.replace(" ",""))

    vowel_count = 0
    consonant_count = 0

    for char in text.lower():

        if char.isalpha():

            if char in vowel:
                vowel_count += 1
            else:
                consonant_count += 1

    return {
        "word_count": len(words),
        "char_count": len(text.replace(" ", "")),
        "vowel_count":vowel_count,
        "consonant_count": consonant_count,
        "longest_word": max(words, key=len),
        "shortest_word": min(words, key=len)
    }


print(sentence_analyser("Python is a powerful programming language"))
```

```python
# Converter Toolkit

def km_to_miles(km):
    return km * 0.621371

def kg_to_pounds(kg):
    return kg * 2.20462

def litres_to_gallons(litres):
    return litres * 0.264172


while True:
    print("\nConverter Toolkit")
    print("1. KM to Miles")
    print("2. KG to Pounds")
    print("3. Litres to Gallons")
    print("4. Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        km = float(input("Enter KM: "))
        print(f"Miles: {km_to_miles(km):.2f}")

    elif choice == "2":
        kg = float(input("Enter KG: "))
        print(f"Pounds: {kg_to_pounds(kg):.2f}")

    elif choice == "3":
        litres = float(input("Enter Litres: "))
        print(f"Gallons: {litres_to_gallons(litres):.2f}")

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice. Try again.")
```

```python
# Advanced Mini Report Card System

def get_student_info():
    name = input("Enter student name: ")

    marks = []

    print("\nEnter marks for 5 subjects:")

    for i in range(1, 6):
        while True:
            mark = float(input(f"Subject {i} marks: "))

            # Validation
            if 0 <= mark <= 100:
                marks.append(mark)
                break
            else:
                print("Marks should be between 0 and 100.")

    return name, marks


def calculate_stats(marks):
    total = sum(marks)
    average = total / len(marks)
    highest = max(marks)
    lowest = min(marks)

    return {
        "total": total,
        "average": average,
        "highest": highest,
        "lowest": lowest
    }

def determine_grade(average):

    if average >= 90:
        return "A+"
    elif average >= 80:
        return "A"
    elif average >= 70:
        return "B"
    elif average >= 60:
        return "C"
    elif average >= 40:
        return "D"
    else:
        return "Fail"

def check_result(marks):

    for mark in marks:
        if mark < 35:
            return "Fail"

    return "Pass"

def print_report(name, marks, stats, grade, result):

    print("\n" + "=" * 40)
    print("         STUDENT REPORT CARD")
    print("=" * 40)

    print(f"Student Name : {name}")

    print("\nSubject Marks:")
    for i, mark in enumerate(marks, start=1):
        print(f"Subject {i} : {mark}")

    print("\Statistics")
    print(f"Total Marks   : {stats['total']}")
    print(f"Average Marks : {stats['average']:.2f}")
    print(f"Highest Marks : {stats['highest']}")
    print(f"Lowest Marks  : {stats['lowest']}")

    print("\nFinal Result")
    print(f"Grade  : {grade}")
    print(f"Result : {result}")


    if result == "Fail":
        print("Performance : Needs Improvement")
    elif grade == "A+":
        print("Performance : Outstanding")
    elif grade == "A":
        print("Performance : Excellent")
    elif grade == "B":
        print("Performance : Very Good")
    else:
        print("Performance : Good")

    print("=" * 40)



print("Welcome to Mini Report Card System")

name, marks = get_student_info()

stats = calculate_stats(marks)

grade = determine_grade(stats["average"])

result = check_result(marks)

print_report(name, marks, stats, grade, result)
```
