```python
#1.Create variables and print using f-string

name = "Ankita"
age = 22
height = 5.58
coffee_preference = "Cold Coffee"

print("My name is {name}") #basic fstring
print(f" Next year I will be {age + 1}") #embedding expression
print(f"height: {height:1f}") #formating decimal 
print(f"|{coffee_preference:^10}|") #padding
```

```python
#2.user input and arithmatic operation
a = int(input("enter the value of a"))
b = int(input("enter the value of b"))

sum = a + b
diff = a - b
product = a * b
result1 = a / b
result2 = a // b
result3 = a % b
print(sum)
print(diff)
print(product)
print(result1)
print(result2)
print(result3)
```

```python
#4.celcius to farenheit
C = float(input("Enter temperature in celcius"))

F = (C * 9/5) + 32
print(f"Temperature in Fahrenheit {F:.1f}")
       
```

```python
#5.Case conversion
W = input("Enter your faviorate word")

print(f"Length {len(W)}")
print(f"Uppercase {W.upper}")
print(f"Lowercase {W.lower}")
print(f"f_char {W[0]}")
print(f"f_char {W[-1]}")
```

```python
#6.billing

item_name = input("Enter item name: ")
price = float(input("Enter price of item: "))
quantity = int(input("Enter quantity: "))

subtotal = price * quantity
tax = subtotal * 0.18
total = subtotal + tax

print("Item:", item_name)
print("Price: ₹", format(price, ".2f"))
print("Quantity:", quantity)
print("Subtotal: ₹", format(subtotal, ".2f"))
print("Tax (18%): ₹", format(tax, ".2f"))
print("Total: ₹", format(total, ".2f"))
```

```python
#7.average,minimum and maximum
a=int(input("enter the test1 score"))
b=int(input("enter the test2 score"))
c=int(input("enter the test3 score"))

average = (a + b + c) / 3
minimum = min(a,b,c)
maximum = max(a,b,c)

print("Average:",average)
print("Minimum:",minimum)
print("Maximum:",maximum)
```
