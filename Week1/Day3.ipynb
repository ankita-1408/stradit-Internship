```python
# Age Category Classifier
age = int(input("Enter your age: "))

if age >= 0 and age <= 12:
    category = "Child"

elif age >= 13 and age <= 17:
    category = "Teenager"

elif age >= 18 and age <= 24:
    category = "Young Adult"

elif age >= 25 and age <= 59:
    category = "Adult"

elif age >= 60 and age <= 150:
    category = "Senior"

else:
    category = "Invalid age"

print("Category:", category)
```

```python
# Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /, %, **): ")

if operator == "+":
    print("Result =", num1 + num2)

elif operator == "-":
    print("Result =", num1 - num2)

elif operator == "*":
    print("Result =", num1 * num2)

elif operator == "/":
    if num2 != 0:
        print("Result =", num1 / num2)
    else:
        print("Cannot divide by zero")

elif operator == "%":
    print("Result =", num1 % num2)

elif operator == "**":
    print("Result =", num1 ** num2)

else:
    print("Unknown operator")
    
```

```python
# Multiplication Table Generator

num = int(input("Enter a number: "))

for i in range(1, 13):
    print(f"{num} x {i} = {num * i}")
```

```python
#PATTERNS
#Pattern A - Right Traingle
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
#Pattern B -Inverted Right Triangle
for i in range(5, 0, -1):
    for j in range(i, 0, -1):
        print(j, end=" ")
    print()
```

```python
# Digit Sum and Product Calculator

num = int(input("Enter a number: "))

sum_digits = 0
product_digits = 1

while num > 0:
    digit = num % 10
    sum_digits += digit
    product_digits *= digit
    num = num // 10

print("Sum of digits =", sum_digits)
print("Product of digits =", product_digits)
```

```python
# Password Strength Checker

password = input("Enter password: ")

has_digit = False
has_upper = False

for ch in password:

    if ch.isdigit():
        has_digit = True

    if ch.isupper():
        has_upper = True

length = len(password)

if length < 6:
    strength = "Weak"

elif length >= 6 and length <= 9:
    if has_digit and has_upper:
        strength = "Medium"
    else:
        strength = "Weak"

else:
    if has_digit and has_upper:
        strength = "Strong"
    else:
        strength = "Medium"

print("Password Strength:", strength)
```
